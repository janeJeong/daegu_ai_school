```python
import math
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

## 일차함수
* $y = ax + b$
+ $a$: 기울기, $b$: y절편
* 그래프 상에서 직선인 그래프(linear)

```python
def linear_function(x):
    a = 0.5
    b = 2
    
    return a * x + b
```

```python
print(linear_function(5))
```

```python
x = np.arange(-5, 5, 0.1)
```

```python
y = linear_function(x)
plt.plot(x, y)
```

## 이차함수
* $y = ax^2 + bx + c$
* 일반적으로 두 개의 실근을 가짐

```python
def quadratic_function(x):
    a = 1
    b = -1
    c = -2
    
    return a*x**2 + b*x + c
```

```python
x = np.arange(-5, 5, 0.1)
y = quadratic_function(x)

plt.plot(x, y)
```

## 삼차함수(다항함수)

* $y = ax^3 + bx^2 + cx + d$

```python
def cubic_function(x) :
    a = 4
    b = 0
    c = -1
    d = -8
    
    return a*x**3 + b*x**2 + c*x + d
```

```python
x = np.arange(-5,5,0.1)
y = cubic_function(x)

plt.plot(x, y)
```

```python
def my_func(x):
    a = 1
    b = -3
    c = 10
    return a*x**2 + b*x + c
```

```python
x = np.arange(-10, 10, 0.1)
y = my_func(x)

min_val = min(y)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.text(0, 20, 'min value of f(x)={}'.format(min_val))

print(min_val)
```

```python
def get_minimum(x1, x2, f):
    x = np.arange(x1, x2, 0.01)
    y = f(x)

    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('get_minimum')
    plt.show()

    return min(y)
```

```python
print(get_minimum(1, 4, my_func))
```
