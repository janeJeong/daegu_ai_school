linspace( ) 함수는 파이썬의 numpy 모듈에 포함된 함수로서 1차원의 배열 만들기, 그래프 그리기에서 수평축의 간격 만들기 등에 매우 편리하게 사용할 수 있는 함수입니다. 이름에서 알 수 있듯이 Linearly Spaced의 줄임말인 것으로 생각되며, 이 시리즈의 [P026]편에서 간단히 언급한 적이 있는 함수입니다.

 사용법은 numpy 모듈을 importing 한 후에 x=np.linspace(start, stop, num)과 같이 적으면 됩니다. start는 배열의 시작값, stop은 배열의 끝값이고, num은 start와 stop 사이를 몇 개의 일정한 간격으로 요소를 만들 것인지를 나타내는 것입니다. 만일 num을 생략하면 디폴트(Default)로 50개의 수열, 즉 1차원 배열을 만들어줍니다.

```python
x = np.linspace(0, 10, 100)   # -> 넘파이 배열, 0에서 10사이에 100개의 값
```

# 간단한 라인 플룻

np.linspace()에 대한 참조 링크 ([https://m.blog.naver.com/choi_s_h/221730568009](https://m.blog.naver.com/choi_s_h/221730568009))
linspace() 함수는 파이썬의 NumPy에 포함된 함수로 1차원의 배열 만들기 혹은 그래프 그리기에서 수평축의 간격 만들기 등에 매우 편리하게 사용할 수 있는 함수 있다. 사용방법은 x = np(staar, stop, num)과 같은 형태로 사용된다. 이때 num 파라메터를 생략하게 되면 기본적으로 50개의 수열을 생성해 준다.

```python
plt.style.use('seaborn-whitegrid')
```

```python
fig = plt.figure()   # -> 그림판 생성
ax = plt.axes()     # -> 축 생성

x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x))
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

## 선 색상과 스타일

```python
plt.plot(x, np.sin(x - 0), color = 'blue')    # 색상을 이름으로 지정
plt.plot(x, np.sin(x - 1), color = 'k')       # 짧은 색상 코드
plt.plot(x, np.sin(x - 2), color = '0.5')     # 0-1 사이의 회색조를 지정 1- 흰색, 단, 문자로 표현해줘야함.
plt.plot(x, np.sin(x - 3), color = '#FFDD44') # RGB (제일 앞에 FF자리가 빨간색 자리(FF-16 진수에서 가장 큰 숫자), DD자리는 녹색(FF보다는 작은 수),44자리는 파란색)
plt.plot(x, np.sin(x - 4), color = (1.0, 0.2, 0.3))  # RGB값을 튜플로 표현 (1.0이 제일 큰 수)
plt.plot(x, np.sin(x - 5), color = 'chartreuse')  # HTML에서 정해진 색상 이름
```

```python
plt.plot(x, x + 0, linestyle = 'solid')  # solid - 실선
plt.plot(x, x + 1, linestyle = 'dashed')
plt.plot(x, x + 2, linestyle = 'dashdot')
plt.plot(x, x + 3, linestyle = 'dotted')

plt.plot(x, x + 4, linestyle = '-')
plt.plot(x, x + 5, linestyle = '--')
plt.plot(x, x + 6, linestyle = '-.')
plt.plot(x, x + 7, linestyle = ':')
```

```python
plt.plot(x, x + 0, '-g')
plt.plot(x, x + 1, '--c')  # 'c' - 시안, 마젠타
plt.plot(x, x + 2, '-.k')
plt.plot(x, x + 3, ':r')
```

## 플롯 조정하기: 축 경계

축 경계를 세밀하게 제어 할 때 사용할 수 있는 함수가 xlim(), ylim()이 있다.
                                                    -> limit

```python
plt.plot(x, np.sin(x))

# 원근을 준다고도 말함
plt.xlim(-1, 11)  # -> 범위를 키운다 = 그래프가 멀어지면서 작아진다
plt.ylim(-1.5, 1.5)
```

```python
plt.plot(x, np.sin(x))

plt.xlim(10, 0)
plt.ylim(1.2, -1,2)    # => 그래프가 뒤집힌 형태로 출력됨
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight')    # -> 그래프를 꽉차게 그림
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5])   # -> 축지정 -1에서 11까지, -1.5에서 1.5까지
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal')    # -> x축 y축의 범위를 동일하게?
```

## 플롯에 레이블 붙이기

제목과 x, y 축에 레이블 그리고 범례를 붙이는 방법을 살펴본다.

```python
plt.plot(x, np.sin(x))
plt.title('A sine Curve')  # plt.title - 표에 제목 붙이기
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
plt.plot(x, np.sin(x), '-g', label = 'sin(x)')
plt.plot(x, np.cos(x), ':b', label = 'cos(x)')
plt.axis('equal')
plt.legend()   # .legend() - 지정한 라벨 값을 확인할 수 있음
```

## 산점도의 출력

라인 플룻에 이어서 점으로 표시하는 산점도(Scatter plot)는 점이나 다른 도형으로 표현한다.

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color = 'black')   # 'o' - 산점도를 나타내는 굵은 점이 나온다
```

```python
x
```

```python
rng = np.random.RandomState(0)
for marker in ['o', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd'] :
    plt.plot(rng.rand(5), rng.rand(5), marker, label = 'marker =' + marker)
    plt.legend()
```

```python
rng.rand(5)
```

```python
rng.rand(5)
```

```python
plt.plot(x, y, '-ok')  # 'ok' - 첫 번째 글자는 산점도, 두 번째 글자는 컬러 / -붙이면 선으로 연결된 산점도가 나타남
```

```python
plt.plot(x, y, '-p', color = 'gray',   # p - pentagon(미국방성), 오각형
        markersize = 15, linewidth = 4,
        markerfacecolor = 'white',
        markeredgewidth = '2')
```

```python
plt.scatter(x, y, marker = 'o')
```

# plt.scatter를 이용한 산점도의 표현

```python
rng = np.random.RandomState(0)
x = rng.random(100)
y = rng.random(100)
color = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c = color, s = sizes, alpha = 0.3, cmap = 'viridis')  # color - 색상 랜덤으로 입히기, size = 원 사이즈 다르게, alpha - 투명도설정
plt.colorbar()  # .colorbar() - 오른쪽에 있는 컬러바 나타냄
```

```python
x
```

```python
from sklearn.datasets import load_iris     # scikik-learn
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], s = 100 * features[3],
           alpha = 0.2, c = iris.target,
           cmap = 'viridis')  # s = size, c = color
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```

# 오차 시각화 하기

```python
x = np.linspace(0, 10, 50)
dy = 0.8   # dy = 오차범위
y = np.sin(x) + dy + np.random.randn(50)
```

```python
plt.errorbar(x, y, yerr = dy, fmt = '.k')
```

```python
plt.errorbar(x, y, yerr = dy, fmt = 'o', color = 'black',
            ecolor = 'lightgray', elinewidth = 3)      # elinewidth - 폭
```
